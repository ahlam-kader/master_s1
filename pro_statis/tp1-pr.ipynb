```python
#7
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def simulate_distribution(distribution, n=1000, params={}):
    if distribution == 'binomiale':
        data = np.random.binomial(n=params.get('n', 10), p=params.get('p', 0.5), size=n)
    elif distribution == 'poisson':
        data = np.random.poisson(lam=params.get('lambda', 3), size=n)
    elif distribution == 'gomtrique':
        data = np.random.geometric(p=params.get('p', 0.5), size=n)
    elif distribution == 'exponentielle':
        data = np.random.exponential(scale=params.get('scale', 1), size=n)
    elif distribution == 'normale':
        data = np.random.normal(loc=params.get('mu', 0), scale=params.get('sigma', 1), size=n)
    else:
        raise ValueError("Distribution non supporte")
    return data


  
def repeat_simulations(distribution, params, n_simulations=100, n_samples=1000):
    means = []  # Pour stocker les moyennes de chaque simulation
    variances = []  # Pour stocker les variances de chaque simulation
    
    for _ in range(n_simulations):
        data = simulate_distribution(distribution, n=n_samples, params=params)  # Simuler les donnes
        means.append(np.mean(data))  # Calculer la moyenne
        variances.append(np.var(data))  # Calculer la variance
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.hist(means, bins=20, color='lightblue', alpha=0.7)
    plt.title(f"Distribution des moyennes\n({distribution})")
    plt.xlabel("Moyenne")
    plt.ylabel("Frquence")
    
    plt.subplot(1, 2, 2)
    plt.hist(variances, bins=20, color='lightgreen', alpha=0.7)
    plt.title(f"Distribution des variances\n({distribution})")
    plt.xlabel("Variance")
    plt.ylabel("Frquence")
    
    plt.tight_layout()
    plt.show()

distributions = {
    'binomiale': {'n': 8, 'p': 0.5},
    'poisson': {'lambda': 2},
    'gomtrique': {'p': 0.5},
    'exponentielle': {'scale': 1},
    'normale': {'mu': 0, 'sigma': 1}
}

for dist, params in distributions.items():
    data = simulate_distribution(dist, n=1000, params=params)      
    repeat_simulations(dist, params, n_simulations=100, n_samples=1000)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataframe = pd.read_csv(r"C:\Users\ahlam\Downloads\Data.csv", sep=";")

print(dataframe.head())
print(dataframe.dtypes)  
```

```python
#1
def categoriser_colonnes(dataframe: pd.DataFrame) -> pd.DataFrame:
    resultats = [{'variable': colonne, 'categorie': 'discrte'} for colonne in dataframe.columns]
    return pd.DataFrame(resultats)

if dataframe.empty:
    print("Erreur : Le fichier CSV est vide.")
    exit()

resultats_colonnes = categoriser_colonnes(dataframe)

print(resultats_colonnes.to_string(index=False))
```

```python
#2
quantitative_vars = dataframe.select_dtypes(include=[np.number])
for col in quantitative_vars.columns:
    mean_value = dataframe[col].mean()
    median_value = dataframe[col].median()
    print(f"{col} - Moyenne: {mean_value:.2f}, Mdiane: {median_value:.2f}")
```

```python
#3
for col in quantitative_vars.columns:
    variance = dataframe[col].var()
    std_dev = dataframe[col].std()
    print(f"{col} - Variance: {variance:.2f}, cart-type: {std_dev:.2f}")
```

```python
#4
resultats = []
for variable in quantitative_vars:
    min_val = dataframe[variable].min()
    max_val = dataframe[variable].max()
    q1 = dataframe[variable].quantile(0.25)
    q2 = dataframe[variable].quantile(0.50)  
    q3 = dataframe[variable].quantile(0.75)
    resultats.append({'variable': variable, 'min': min_val, 'Q1': q1, 'Q2 (mdiane)': q2, 'Q3': q3, 'max': max_val})

stats = pd.DataFrame(resultats)
print(stats)
```

```python
#5
for col in quantitative_vars.columns:
    Q1 = dataframe[col].quantile(0.25)
    Q3 = dataframe[col].quantile(0.75)
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)]
    print(f"{col} - Nombre de valeurs extrmes: {outliers.shape[0]}")
```

```python
#8
from scipy.stats import skew, kurtosis

for col in quantitative_vars:
    skewness = skew(dataframe[col].dropna())
    kurt = kurtosis(dataframe[col].dropna())
    
    print(f"{col} - Asymtrie: {skewness:.2f}, Kurtosis: {kurt:.2f}")
    
    plt.figure(figsize=(8, 5))
    plt.hist(dataframe[col].dropna(), bins=30, alpha=0.6, color='lightblue', density=True, label=f'Histogramme de {col}')
    plt.title(f"Distribution de {col}")
    plt.xlabel(col)
    plt.ylabel('Densit')
    plt.legend()
    plt.show()
```

```python
#9
from scipy.stats import shapiro, norm  

for col in quantitative_vars:
    sample = dataframe[col].dropna().sample(n=5000, random_state=42) 
    stat, p = shapiro(sample) 

    if p > 0.05:
        print(f"{col} : Distribution normale probable (p-value: {p:.4f})")
    else:
        print(f"{col} : Distribution non normale (p-value: {p:.4f})")
    
    plt.figure(figsize=(10, 6))
    plt.hist(sample, bins=30, alpha=0.6, color='lightblue', density=True, label='Donnes')

    mu, std = np.mean(sample), np.std(sample)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p_norm = norm.pdf(x, mu, std) 
    plt.plot(x, p_norm, 'k', linewidth=2, label='Normale thorique')

    plt.title(f'{col} - Histogramme et Courbe Normale')
    plt.xlabel(col)
    plt.ylabel('Densit')
    plt.legend()
    plt.show()
```
